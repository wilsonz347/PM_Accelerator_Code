# Data Preprocessing

```python
# Narrow the forecasting to a specific country
countries = df['country'].unique()

country_to_model = 'United States of America' 

# Filter data for the chosen country
country_df = df[df['country'] == country_to_model].copy()

# Convert the last_updated column to datetime object
country_df['last_updated'] = pd.to_datetime(country_df['last_updated'])
country_df = country_df.set_index('last_updated')

country_df = country_df.sort_index()
    
# Drop columns that are irrelevant and redundant 
columns_to_drop = [
    'country', 'location_name', 'timezone', 'longitude', 'latitude',
    'temperature_fahrenheit', 'feels_like_celsius', 'feels_like_fahrenheit',  
    'pressure_in', 'precip_in', 'visibility_miles', 'gust_mph', 'wind_mph',
    'condition_text', 'wind_direction', 'last_updated_epoch', 
    'sunrise', 'sunset', 'moonrise', 'moonset', 'moon_phase', 'moon_illumination',  
    'air_quality_Carbon_Monoxide', 'air_quality_Ozone', 'air_quality_Nitrogen_dioxide', 
    'air_quality_Sulphur_dioxide', 'air_quality_PM2.5', 'air_quality_PM10', 
    'air_quality_us-epa-index', 'air_quality_gb-defra-index', 'visibility_km', 'cloud', 
    'precip_mm', 'uv_index'
]

country_df = country_df.drop(columns=columns_to_drop)
country_df = country_df.dropna().drop_duplicates()
print(f"Shape: {country_df.shape}")

# Handle Outliers by IQR
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

country_df = handle_outliers(country_df, 'temperature_celsius')
print(f"New size: {len(country_df)}")

# Resample
country_df = country_df.resample('D').mean() 
country_df = country_df.ffill()

# Verify the resampled data
display(country_df.head())
print(f"Shape after resampling: {country_df.shape}")
```

```python
# Save to csv file
country_df.to_csv('filtered_data.csv', index=False)
```

```python
country_df.dtypes
```

```python
from pmdarima import auto_arima

target = country_df['temperature_celsius']
features = country_df.drop('temperature_celsius', axis=1)

# First-order differencing
target_diff = target.diff().dropna()

# Split the data  
training_size = int(len(features) * 0.8) 
train_target, test_target = target[:training_size], target[training_size:]
train_features, test_features = features[:training_size], features[training_size:]

# Library to automate the parameters
model = auto_arima(
    train_target,               
    exogenous=train_features,       
    m=12,               
    stepwise=True,    
    suppress_warnings=True
)
```

# Model Training

```python
# SARIMA model 
from statsmodels.tsa.statespace.sarimax import SARIMAX
order = model.order            
seasonal_order = model.seasonal_order

SARIMA_model = SARIMAX(train_target, exog=train_features, order=order, seasonal_order=seasonal_order)
SARIMA_model_fit = SARIMA_model.fit()

print(SARIMA_model_fit.summary())
```

```python
# Plot the residual
residuals = SARIMA_model_fit.resid[1:]

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
residuals.plot(title='Residual of model')
plt.xlabel('Time')
plt.ylabel('Residuals')
plt.grid(True)

plt.subplot(1,2,2)
residuals.plot(title='Density of Residuals', kind='kde')
plt.xlabel('Residuals')
plt.grid(True)

plt.show()
```

### Interpretation of the residual plots
- The residual plot seems to be randomly distributed, as indicated by the oscillation. 
- The density plot appears to be bell-shaped. 
    - The peak is centered near 0, but slightly shifted to the left.
    
Conclusion:
- The residuals seems to be normally distributed.
- The range of the residual plot indicates that the model's prediction fluctuates.

```python
# Make predictions on the test set
forecast = SARIMA_model_fit.forecast(steps=len(test_target), exog=test_features)

# Calculate error metrics
from sklearn.metrics import mean_squared_error
import numpy as np

rmse = np.sqrt(mean_squared_error(test_target, forecast))
print(f"RMSE: {rmse}")

# Plot actual vs predicted plot
plt.figure(figsize=(12,6))
plt.plot(test_target.index, test_target, label='Actual')
plt.plot(test_target.index, forecast, label='Forecast')
plt.title('Temperature Forecast: Actual vs Predicted')
plt.xlabel('Date')
plt.ylabel('Temperature (Celsius)')
plt.legend()
plt.grid(True)

plt.show()
```
